```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../data/incidents_train.csv')

data_info = data.info()
data_head = data.head()
data_info, data_head
```

```python
# Compute lengths of titles
data['title_length'] = data['title'].apply(len)

# Plot histogram of title lengths
plt.figure(figsize=(10, 6))
plt.hist(data['title_length'], bins=10, edgecolor='black', alpha=0.7)
plt.title('Histogram of Title Lengths')
plt.xlabel('Title Length')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data['hazard-category'].value_counts().plot(kind='bar')
plt.title('Distribution of Hazard Categories')
plt.xlabel('Hazard Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
data['product-category'].value_counts().plot(kind='bar')
plt.title('Distribution of Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
data.groupby('year').size().plot(kind='line', marker='o')
plt.title('Number of Incidents Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data['hazard'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Most Frequent Specific Hazards')
plt.xlabel('Hazard')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

top_product_per_category = data.groupby(['product-category', 'product']).size().reset_index(name='count')
top_products = top_product_per_category.sort_values(by='count', ascending=False).groupby('product-category').head(3)

plt.figure(figsize=(12, 8))
sns.barplot(data=top_products, y='product', x='count', hue='product-category', dodge=False)
plt.title('Top Products by Category')
plt.xlabel('Frequency')
plt.ylabel('Product')
plt.xticks(rotation=0)
plt.yticks(rotation=0, ha='right')
plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()


data['date'] = pd.to_datetime(data[['year', 'month', 'day']])
monthly_trend = data.groupby([data['date'].dt.year, data['date'].dt.month]).size().unstack(fill_value=0)

plt.figure(figsize=(12, 8))
sns.heatmap(monthly_trend, cmap="YlGnBu", annot=True, fmt="d", cbar_kws={'label': 'Incident Count'})
plt.title('Monthly Trend of Incidents Over Years')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()
```

```python
top_10_products = top_product_per_category.sort_values(by='count', ascending=False).head(10)

# Plotting the top 10 products by frequency with category hue
plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_products, y='product', x='count', hue='product-category', dodge=False)
plt.title('Top 10 Products by Category')
plt.xlabel('Frequency')
plt.ylabel('Product')
plt.xticks(rotation=0)
plt.yticks(rotation=0, ha='right')
plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
top_hazard_per_category = data.groupby(['hazard-category', 'hazard']).size().reset_index(name='count')
top_10_hazards = top_hazard_per_category.sort_values(by='count', ascending=False).head(10)

# Plotting the top 10 hazards by frequency with category hue
plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_hazards, y='hazard', x='count', hue='hazard-category', dodge=False)
plt.title('Top 10 Hazards by Category')
plt.xlabel('Frequency')
plt.ylabel('Hazard')
plt.xticks(rotation=0)
plt.yticks(rotation=0, ha='right')
plt.legend(title='Hazard Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
